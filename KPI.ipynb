```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("Camp_Market_nettoye.csv", sep=';')
df['Age'] = 2024 - df['Year_Birth']
df['TotalChildren'] = df['Kidhome'] + df['Teenhome']
df['TotalSpent'] = df[['MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds']].sum(axis=1)
df['TotalPurchases'] = df[['NumWebPurchases','NumCatalogPurchases','NumStorePurchases','NumDealsPurchases']].sum(axis=1)
df['AcceptedCampaigns'] = df[['AcceptedCmp1','AcceptedCmp2','AcceptedCmp3','AcceptedCmp4','AcceptedCmp5']].sum(axis=1)
```

```python
le = LabelEncoder()
df['Education_encoded'] = le.fit_transform(df['Education'])
df['Marital_encoded'] = le.fit_transform(df['Marital_Status'])
```

```python

# Configuration des graphiques
sns.set(style="whitegrid")
plt.rcParams.update({'figure.figsize':(10,6), 'figure.dpi':100})

# Histogramme de l'âge
sns.histplot(df['Age'], bins=30, kde=True, color='skyblue')
plt.title("Distribution de l'âge des clients")
plt.xlabel("Âge")
plt.ylabel("Nombre de clients")
plt.show()

# Boxplot des dépenses par réponse
sns.boxplot(x='Response', y='TotalSpent', data=df, palette='Set2')
plt.title("Dépenses totales selon la réponse à la campagne")
plt.xlabel("Réponse à la campagne")
plt.ylabel("Dépenses totales (€)")
plt.show()

# Countplot campagnes acceptées
sns.countplot(x='AcceptedCampaigns', data=df, palette='Set3')
plt.title("Nombre de campagnes acceptées par client")
plt.xlabel("Campagnes acceptées")
plt.ylabel("Nombre de clients")
plt.show()

# Scatter Age vs Dépenses
sns.scatterplot(x='Age', y='TotalSpent', hue='Response', data=df, palette='coolwarm')
plt.title("Âge vs Dépenses totales (coloré par réponse)")
plt.xlabel("Âge")
plt.ylabel("Dépenses totales (€)")
plt.show()

# Visites web
sns.histplot(df['NumWebVisitsMonth'], bins=10, color='orange')
plt.title("Fréquence des visites web par mois")
plt.xlabel("Visites web par mois")
plt.ylabel("Nombre de clients")
plt.show()

# Achats par canal
channels = ['NumWebPurchases','NumCatalogPurchases','NumStorePurchases','NumDealsPurchases']
sns.barplot(x=channels, y=df[channels].sum().values, palette='Blues')
plt.title("Nombre total d'achats par canal")
plt.xlabel("Canal d'achat")
plt.ylabel("Nombre total d'achats")
plt.show()

# Réponse par statut marital
sns.barplot(x='Marital_Status', y='Response', data=df, estimator=np.mean, palette='Purples')
plt.title("Taux de réponse par statut marital")
plt.xlabel("Statut marital")
plt.ylabel("Taux moyen de réponse")
plt.show()

# Réponse par niveau d'éducation
sns.barplot(x='Education', y='Response', data=df, estimator=np.mean, palette='Greens')
plt.title("Taux de réponse par niveau d'éducation")
plt.xlabel("Niveau d'éducation")
plt.ylabel("Taux moyen de réponse")
plt.show()

# Dépenses moyennes par produit
products = ['MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds']
sns.barplot(x=products, y=df[products].mean().values, palette='Reds')
plt.title("Dépenses moyennes par type de produit")
plt.xlabel("Type de produit")
plt.ylabel("Dépenses moyennes (€)")
plt.show()

# Plaintes clients
sns.countplot(x='Complain', data=df, palette='pastel')
plt.title("Analyse des plaintes clients")
plt.xlabel("Plaintes (0 = Non, 1 = Oui)")
plt.ylabel("Nombre de clients")
plt.show()
```
